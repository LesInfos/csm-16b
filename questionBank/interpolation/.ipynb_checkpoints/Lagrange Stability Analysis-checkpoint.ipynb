```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
def generate_polynomial(coeff):
    n = len(list(coeff))
    return lambda x: sum([coeff[i] * x**i for i in range(n)])
```

```python
ct_space = np.linspace(0, 10, 1000)
ct_signal = 3/2 - 1/2 * np.cos(np.pi * ct_space)

plt.figure(figsize=(10,4))
plt.plot(ct_space, ct_signal)
plt.title('Continuous signal: $3/2 - 1/2 \cos(\pi t)$');
plt.xlabel('t')
plt.ylabel('x(t)')
plt.savefig('figures/ct_signal.png', dpi=600);
```

```python
two_pts = np.arange(0, 2)
two_samples = 3/2 - 1/2 * np.cos(np.pi * two_pts)

coeff = lagrange(two_pts, two_samples)

f = generate_polynomial(coeff)

two_space = np.linspace(0, 1, 100)

plt.figure(figsize=(7,4))
plt.scatter(two_pts, two_samples)
plt.title('Two samples from the original signal');
plt.xlabel('t')
plt.ylabel('x(t)')
plt.savefig('figures/two_pt_samples.png', dpi=600);

plt.figure(figsize=(7,4))
plt.plot(two_space, f(two_space))
plt.scatter(two_pts, two_samples)
plt.title('Two samples from the original signal');
plt.xlabel('t')
plt.ylabel('x(t)')
plt.savefig('figures/two_pt_lagrange.png', dpi=600);
```
